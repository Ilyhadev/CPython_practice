## Задание 1

Реализуйте класс с полностью инкапсулированным состоянием, используя name mangling и property, обеспечив валидацию при изменении атрибутов и демонстрируя, как Python скрывает "приватные" данные.

```python
class Encapsulated:
    def __init__(self, value):
        self.__value = None
        self.value = value

    @property
    def value(self):
        return self.__value

    @value.setter
    def value(self, new_value):
        # простой пример валидации
        if not isinstance(new_value, (int, float)):
            raise TypeError("Value must be a number")
        if new_value < 0:
            raise ValueError("Value must be non-negative")
        
        self.__value = new_value

    @value.deleter
    def value(self):
        print("Deleting value...")
        del self.__value

obj = Encapsulated(10)
print(obj.value)
print(obj._Encapsulated__value)

# obj2 = Encapsulated(-123)
# print(obj2.value)
# print(obj2._Encapsulated__value)
```

Создайте иерархию классов с абстрактным базовым классом (ABC) и абстрактными методами, демонстрируя использование модуля abc. Реализуйте интерфейсы и их проверку через isinstance и issubclass.

```python
from abc import ABC, abstractmethod

# Your code here (for example you can create Animal(ABC) with method sound())

class Animal(ABC):
    """Абстрактный базовый класс для всех животных"""
    
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def sound(self):
        pass



class Dog(Animal):
    def sound(self):
        return "Bark"


class Cat(Animal):
    def sound(self):
        return "Meow"
    


class Bird(Animal):
    def sound(self):
        return "Chirp"


dog = Dog("Sharik")
cat = Cat("Barsik")
bird = Bird("Bird")

print(dog.sound())
print(cat.sound())
print(bird.sound())

print(f"  dog is Animal? {isinstance(dog, Animal)}")
print(f"  dog is Dog? {isinstance(dog, Dog)}")
print(f"  dog is Cat? {isinstance(dog, Cat)}")
```

## Задание 5

Создайте класс, который использует метакласс, объясните как метаклассы влияют на создание классов в Python, и реализуйте контролируемое изменение класса через метакласс.

```python
class Meta(type):
    def __new__(cls, name, bases, namespace):
        namespace['version'] = '1.0'
        if 'process' not in namespace:
            print(f"class {name} has no 'process' method")

        return super().__new__(cls, name, bases, namespace)

class GoodClass(metaclass=Meta):
    def process(self):
        print("Processing...")

class BadClass(metaclass=Meta):
    pass

print(GoodClass.version)
print(BadClass.version)

# Basically metaclass is a fabric OOP design pattern
```

## Задание 6

Реализуйте класс с дескрипторами данных и неданных, объясните разницу между ними и механизм вызова методов get, set, delete, особенно при наследовании.

```python
class DataDescriptor:
    def __get__(self, obj, objtype=None):
        print(f"__get__")
        return 42 if obj else self
    
    def __set__(self, obj, value):
        print(f"__set__ to {value}")
        obj.__dict__['data_desc'] = value
    
    def __delete__(self, obj):
        print("__delete__")
        del obj.__dict__['data_desc']

class NonDataDescriptor:
    def __get__(self, obj, objtype=None):
        print(f"NonDataDescriptor __get__")
        return 99 if obj else self

class MyClass2:
    data_desc = DataDescriptor()
    non_data_desc = NonDataDescriptor()

obj = MyClass2()

obj = MyClass2()
obj.data_desc = 10
print(obj.data_desc)
print(obj.non_data_desc)


# Data descriptor (есть __set__ или __delete__): всегда перехватывает доступ, даже если есть атрибут экземпляра
# Non-data descriptor (только __get__): перехватывает доступ только если нет атрибута экземпляра
```

## Задание 8

Реализуйте класс с пользовательскими dunder-методами: new, init, call, del, repr, str, и объясните, как Python вызывает их в цикле жизни объекта.

```python
class LifeCycle:
   value = 10
   def __new__(cls):
        print("__new__ called")
        instance = super().__new__(cls)
        instance.value = 0
        return instance
    
   def __init__(self, value=None):
        print("__init__ called")
        self.value = value
   
   def __call__(self):
      print("Call invoked")
   
   def __repr__(self):
      return f"LifeCycle {self.value} repr"

   def __str__(self):
      return f"LifeCycle {self.value} str"

obj = LifeCycle()
print(obj)
obj()
del obj

# __new__: creates instance, must return it
# __init__: initializes created instance
# __call__: makes instance callable (obj())
# __repr__: representation for debugging
# __str__: string representation for printing
```

## Задание 9

Создайте контекстный менеджер с помощью специальных методов enter и exit, используйте его вместе с классом, в котором присутствуют методы с разграничением прав доступа.

https://habr.com/ru/articles/739326/

```python
class Access:
    def __init__(self):
        self._secret = None
        self.secret = 10
    # Your code here
    def __enter__(self):
        print("entered in context")
        return self._secret
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("exited from context")
        self._secret = 0
    
    @property
    def secret(self):
        return self._secret

    @secret.setter
    def secret(self, new_value):
        self._secret = new_value

obj = Access()
print(obj.secret)
with obj as secret_value:
    print(secret_value)
print(obj.secret)
```

## Задание 10

Создайте класс, объекты которого могут быть отслежены с помощью слабых ссылок (weakref). Реализуйте систему, которая хранит слабые ссылки на все созданные объекты класса и автоматически удаляет их из списка при уничтожении объектов. Продемонстрируйте это поведение, выводя текущее количество живых объектов.

```python
import weakref

class Tracked:
    _instances = []
    
    def __init__(self, name):
        self.name = name
        # добавляем слабую ссылку на себя
        self._instances.append(weakref.ref(self, self._cleanup))
        print(f"Created {self.name}")
    
    @classmethod
    def _cleanup(cls, ref):
        """Автоматически вызывается при удалении объекта"""
        print("cleaning up")
        cls._instances = [r for r in cls._instances if r() is not None]
    
    @classmethod
    def alive_count(cls):
        """Возвращает количество живых объектов"""
        return len([r for r in cls._instances if r() is not None])
    
    def __del__(self):
        print(f"Deleting {self.name}")

# Демонстрация
a = Tracked("A")
b = Tracked("B")
c = Tracked("C")

print(f"Current count: {Tracked.alive_count()}")

print("Deleting B")
del b

print(f"Current count: {Tracked.alive_count()}")

print("Deleting A and C")
del a
del c

print(f"Final count: {Tracked.alive_count()}")
```
